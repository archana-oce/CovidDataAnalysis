```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your COVID-19 dataset (replace 'country_wise_latest.csv' with your file path)
df = pd.read_csv('country_wise_latest.csv')

def display_first_few_rows():
    print(df.head())

def display_summary_statistics():
    print("\nSummary Statistics:")
    print("1. Mean")
    print("2. Median")
    print("3. Mode")
    print("4. Minimum")
    print("5. Maximum")
    print("6. Standard Deviation")
    print("7. Skewness")
    print("8. Kurtosis")
    choice = input("Enter your choice (1-8): ")

    numeric_columns = df.select_dtypes(include='number').columns

    if choice == '1':
        print("\nMean:")
        print(df[numeric_columns].mean())
    elif choice == '2':
        print("\nMedian:")
        print(df[numeric_columns].median())
    elif choice == '3':
        print("\nMode:")
        print(df[numeric_columns].mode().iloc[0]) 
    elif choice == '4':
        print("\nMinimum:")
        print(df[numeric_columns].min())
    elif choice == '5':
        print("\nMaximum:")
        print(df[numeric_columns].max())
    elif choice == '6':
        print("\nStandard Deviation:")
        print(df[numeric_columns].std())
    elif choice == '7':
        print("\nSkewness:")
        print(df[numeric_columns].skew())
    elif choice == '8':
        print("\nKurtosis:")
        print(df[numeric_columns].kurtosis())
    else:
        print("Invalid choice. Please enter a number between 1 and 8.")

def filter_data():
    threshold = int(input("Enter the threshold for confirmed cases: "))
    filtered_df = df[df['Confirmed'] > threshold]
    print("\nFiltered Data:")
    print(filtered_df.head(10))

def group_by_region():
    column = input("Enter the column name to group with: ")
    if column not in df.columns:
        print("Invalid column name.")
        return
    grouped_df = df.groupby('Country/Region')[column].sum()
    print(f"\nTotal {column} by Country/Region:")
    print(grouped_df)


def plot_bar_chart(df):
    column = input("Enter the column name for the bar chart: ")
    plt.figure(figsize=(30, 10))  
    df.groupby("Country/Region")[column].sum().plot(kind='bar')
    plt.xlabel("Country/Region")
    plt.ylabel(column)
    plt.title("Bar Chart")
    plt.xticks(rotation=90) 
    plt.show()

def plot_line_chart(df):
    x_column = input("Enter the column name for the x-axis: ")
    y_column = input("Enter the column name for the y-axis: ")
    df.plot(x=x_column, y=y_column, figsize=(30, 10)) 
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.title("Line Chart")
    plt.show()
    
def plot_scatter_plot(df):
    x_column = input("Enter the column name for the x-axis: ")
    y_column = input("Enter the column name for the y-axis: ")
    df.plot(kind='scatter', x=x_column, y=y_column,figsize=(30,10))
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.title("Scatter Plot")
    plt.show()

def plot_histogram(df):
    column = input("Enter the column name for the histogram: ")
    column_to_plot = column  
    df[column_to_plot].plot(kind='hist',bins=30, figsize=(30, 10))
    plt.xlabel('Value')
    plt.ylabel(column_to_plot)
    plt.title(f'Histogram of {column_to_plot}')
    plt.show()


def plot_pie_chart(df):
    column = input("Enter the column name for the pie chart: ")
    value_counts = df[column].value_counts()
    value_counts.plot(kind='pie', autopct='%1.1f%%', labels=value_counts.index, figsize=(30, 10))
    plt.title("Pie Chart")
    plt.show()

def top_10_data():
    column = input("Enter the column name to show top 10 data: ")
    if column not in df.columns:
        print("Invalid column name.")
        return
    top_10 = df.nlargest(10, column)
    print(f"\nTop 10 data based on {column}:")
    print(top_10)

def plot_data(df):
    print("\nChoose the type of graph:")
    print("1. Bar Chart")
    print("2. Line Chart")
    print("3. Scatter Plot")
    print("4. Histogram")
    print("5. Pie Chart")

    choice = input("Enter your choice (1-5): ")

    if choice == '1':
        plot_bar_chart(df)
    elif choice == '2':
        plot_line_chart(df)
    elif choice == '3':
        plot_scatter_plot(df)
    elif choice == '4':
        plot_histogram(df)
    elif choice == '5':
        plot_pie_chart(df)
    else:
        print("Invalid choice. Please enter a number between 1 and 5.")


def data_analysis():
    print("\nData Analysis Options:")
    print("1. Mortality Rate")
    print("2. Recovery Rate")
    print("3. Active Cases")
    print("4. Case Fatality Ratio")
    
    choice = input("Enter your choice (1-4): ")

    if choice == '1':
        calculate_and_plot_mortality_rate()
    elif choice == '2':
        calculate_and_plot_recovery_rate()
    elif choice == '3':
        calculate_and_plot_active_cases()
    elif choice == '4':
        calculate_and_plot_case_fatality_ratio()
    else:
        print("Invalid choice. Please enter a number between 1 and 4.")

def calculate_and_plot_mortality_rate():
    df['Mortality Rate'] = (df['Deaths'] / df['Confirmed']) * 100
    print("\nMortality Rate calculated.")
    plt.figure(figsize=(30, 6))
    plt.bar(df['Country/Region'], df['Mortality Rate'])
    plt.xlabel('Country/Region')
    plt.ylabel('Mortality Rate (%)')
    plt.title('Mortality Rate by Country/Region')
    plt.xticks(rotation=90)
    plt.show()

def calculate_and_plot_recovery_rate():
    df['Recovery Rate'] = (df['Recovered'] / df['Confirmed']) * 100
    print("\nRecovery Rate calculated.")
    plt.figure(figsize=(30, 6))
    plt.bar(df['Country/Region'], df['Recovery Rate'])
    plt.xlabel('Country/Region')
    plt.ylabel('Recovery Rate (%)')
    plt.title('Recovery Rate by Country/Region')
    plt.xticks(rotation=90)
    plt.show()

def calculate_and_plot_active_cases():
    df['Active Cases'] = df['Confirmed'] - df['Deaths'] - df['Recovered']
    print("\nActive Cases calculated.")
    plt.figure(figsize=(30, 6))
    plt.bar(df['Country/Region'], df['Active Cases'])
    plt.xlabel('Country/Region')
    plt.ylabel('Active Cases')
    plt.title('Active Cases by Country/Region')
    plt.xticks(rotation=90)
    plt.show()

def calculate_and_plot_case_fatality_ratio():
    df['Case Fatality Ratio'] = (df['Deaths'] / (df['Deaths'] + df['Recovered'])) * 100
    print("\nCase Fatality Ratio calculated.")
    plt.figure(figsize=(30, 6))
    plt.bar(df['Country/Region'], df['Case Fatality Ratio'])
    plt.xlabel('Country/Region')
    plt.ylabel('Case Fatality Ratio (%)')
    plt.title('Case Fatality Ratio by Country/Region')
    plt.xticks(rotation=90)
    plt.show()


def calculate_and_plot_deviation():
    deviation = df.std()
    print("\nStandard Deviation:")
    print(deviation)
    deviation.plot(kind='bar', figsize=(30, 10))  
    plt.xlabel('Column')
    plt.ylabel('Standard Deviation')
    plt.title('Standard Deviation of Columns')
    plt.show()

def calculate_and_plot_skewness():
    skew = df.skew()
    print("\nSkewness:")
    print(skew)
    skew.plot(kind='bar', figsize=(30, 10))  
    plt.xlabel('Column')
    plt.ylabel('Skewness')
    plt.title('Skewness of Columns')
    plt.show()

def calculate_and_plot_kurtosis():
    kurt = df.kurtosis()
    print("\nKurtosis:")
    print(kurt)
    kurt.plot(kind='bar', figsize=(30, 10))  
    plt.xlabel('Column')
    plt.ylabel('Kurtosis')
    plt.title('Kurtosis of Columns')
    plt.show()

# Main menu
while True:
    print("\nCOVID-19 Data Analysis Menu:")
    print("1. Display the first few rows")
    print("2. Display summary statistics")
    print("3. Filter data by confirmed cases")
    print("4. Group data by region")
    print("5. Plot data")
    print("6. Show top 10 data")
    print("7. Data Analysis")
    print("8. Exit")

    choice = input("Enter your choice (1-8): ")

    if choice == '1':
        display_first_few_rows()
    elif choice == '2':
        display_summary_statistics()
    elif choice == '3':
        filter_data()
    elif choice == '4':
        group_by_region()
    elif choice == '5':
        plot_data(df)
    elif choice == '6':
        top_10_data()
    elif choice == '7':
        data_analysis()
    elif choice == '8':
        print("Exiting...")
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 8.")
```
